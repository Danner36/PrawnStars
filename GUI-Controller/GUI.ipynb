```python
### IMPORT STATEMENTS ###

from __future__ import print_function
from inputs import get_gamepad
from IPython.display import clear_output
from wireless import Wireless
from subprocess import Popen, STDOUT, PIPE
from time import sleep
```

```python
### GLOBAL VARIABLES ###

#Declared the event object as a global. 
events = 0

#Used to tell if the current button data set is different 
#   than the last.
buttonFlag = 0

#Xbox 1 controller buttons. Turns True when
#   pressed and false otherwise.
A_button = False
B_button = False
X_button = False
Y_button = False


### FUCNTIONS ###

#Cycles through the objects data to check for desired 
#   button presses.
def buttonPress():
    
    #Defines Global Variables. Does not Declare
    global buttonFlag
    global A_button
    global B_button
    global X_button
    global Y_button
    global events
    
    #Gathers the current information from the Xbox 1 Controller.
    events = get_gamepad()
    
    A_button = False
    B_button = False
    X_button = False
    Y_button = False
    
    #Cycles through the objects data to check for desired 
    #   button presses.
    for event in events:

        #Checks if button 'Y' was pressed.
        if((event.code == "BTN_NORTH") and (event.state == 1)):
            #print(event.code, event.state)
            Y_button = True
            buttonFlag = 1

        #Checks if button 'B' was pressed.
        if((event.code == "BTN_EAST") and (event.state == 1)):
            #print(event.code, event.state)
            B_button = True
            buttonFlag = 1

        #Checks if button 'A' was pressed.
        if((event.code == "BTN_SOUTH") and (event.state == 1)):
            #print(event.code, event.state)
            A_button = True
            buttonFlag = 1

        #Checks if button 'X' was pressed.
        if((event.code == "BTN_WEST") and (event.state == 1)):
            #print(event.code, event.state)
            X_button = True
            buttonFlag = 1


def WiFi_Setup():
    print("Enter the name of the desired wifi network. (PUT THE NAME IS QUOTES TO SIGNAL STRING)")
    WiFi_name = str(input())
    connect_to_network(WiFi_name)
    
    print("WiFi Connected. Please verify conneciton to: " + str(WiFi_name))
            
            
            
            
def connect_to_network(name):
    
    handle = Popen('netsh wlan connect ' + str(name), shell=True, stdout=PIPE, stderr=STDOUT, stdin=PIPE)
    
    

#MAIN LOOP 
#   Passively watches the COM ports for information.
#   When detected, reads in the event and its attributes.
def main():
    
    #Defines Global Variables. Does not Declare
    global buttonFlag
    global A_button
    global B_button
    global X_button
    global Y_button
    
    #Walks through the Wifi setup.
    WiFi_Setup()
    
    
    #Inifinte Loop
    while 1:
        
        #Checks for button presses.
        buttonPress()
        
        #Only prints out data when it updates. (upon button press)
        if(buttonFlag == 1):
            
            #Clears the output terminal from last iteration.
            
            
            #Clears the flag.
            buttonFlag = 0
            
            #Prints out the current most values.
            print("Button Values (True = Press)")
            print("A:" + str(A_button))
            print("B:" + str(B_button))
            print("X:" + str(X_button))
            print("Y:" + str(Y_button))
        


#This waits the for the controller is initalize and send its
#   call in value. (Waits for it to start)
if __name__ == "__main__":
   main()
```
